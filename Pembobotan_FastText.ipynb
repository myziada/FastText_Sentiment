```python
import pandas as pd
from gensim.models import FastText

# Read the CSV file
df = pd.read_csv('/content/output_Label_Number_FIX!.csv')

# Extract the sentences from the CSV file
sentences = df['text'].tolist()

# Preprocess the sentences by splitting them into lists of words
sentences = [sentence.split() for sentence in sentences]

# Train the FastText model
model = FastText(sentences, vector_size=100, window=5, min_count=1, workers=4, sg=1)

# Save the trained model to a file
model.save("fasttext_model.model")
```

```python
from gensim.models import FastText

# Load the FastText model
model = FastText.load("fasttext_model.model")

# Iterate over the vocabulary and print word indices
for word, index in model.wv.key_to_index.items():
    print(f"Word: {word}, Index: {index}")
```

```python
similarity_score = model.wv.evaluate_word_pairs('path/to/word_pairs.txt')
print(similarity_score)
```

```python
analogy_score = model.wv.accuracy('path/to/analogies.txt')
print(analogy_score)
```

```python
similar_words = model.wv.most_similar('word')
print(similar_words)
```

```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('/content/1020DataNewABHeader.csv')

# Preprocess the text if needed
df['text'] = df['text'].str.lower()

# Combine label and text into a single column with the required format
df['fasttext_input'] = '__label__' + df['Label'].astype(str) + ' ' + df['text']

# Save the FastText input to a text file
df['fasttext_input'].to_csv('cooking.train', index=False, header=False, sep='\t')


from sklearn.model_selection import train_test_split
import pandas as pd

# Split the data into training and test sets
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Save the training and test sets to separate files
train_data['fasttext_input'].to_csv('cooking.train', index=False, header=False, sep='\t')
test_data['fasttext_input'].to_csv('cooking.test', index=False, header=False, sep='\t')

model = fasttext.train_supervised(input="cooking.train", lr=0.5, epoch=50, wordNgrams=4, bucket=200000, dim=50, loss='hs')

# Save the trained model
model.save_model("modelngram4.bin")

import fasttext

# Load the trained model
model = fasttext.load_model('modelngram4.bin')

# Evaluate the model on the test data
result = model.test('cooking.test')

# Extract the evaluation metrics from the result tuple
precision = result[1]
recall = result[2]
num_examples = result[0]

# Print the evaluation results
print('Precision@1:', precision)
print('Recall@1:', recall)
print('Number of examples:', num_examples)
```
