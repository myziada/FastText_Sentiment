To check if a Word2Vec model has been trained successfully and can be used effectively, you can perform the following checks:

1. Accessing Word Vectors:
You can check if word vectors are accessible for specific words in the vocabulary. This can be done by trying to access the word vector using the model.wv[word] syntax. If the word vector can be accessed without any errors, it indicates that the model has successfully learned word representations.

```python
if 'baswedan' in model.wv:
    print("Word vector for 'word' is accessible.")
else:
    print("Word vector for 'word' is not accessible.")
```

2. Word Similarity Evaluation:
Word2Vec models are often evaluated based on their ability to capture semantic similarity between words. You can use the model.wv.similarity(word1, word2) method to check the similarity between two words. If the similarity score aligns with your expectations (e.g., similar words have higher similarity scores), it suggests that the model has captured meaningful semantic relationships.

```python
word1 = 'anies'
word2 = 'baswedan'
similarity_score = model.wv.similarity(word1, word2)
print(f"Similarity between '{word1}' and '{word2}': {similarity_score}")
```

```python
word1 = 'jalan'
word2 = 'tol'
similarity_score = model.wv.similarity(word1, word2)
print(f"Similarity between '{word1}' and '{word2}': {similarity_score}")
```

3. Word Analogies Evaluation:
Word2Vec models should also perform well on word analogies tasks, such as "king - man + woman = queen." You can use the model.wv.most_similar(positive, negative) method to find the most similar word based on the provided analogy. If the model produces reasonable and expected results for various analogies, it indicates a successful training.

```python
analogy_result = model.wv.most_similar(positive=['janji', 'jalan'], negative=['bohong'])
print(f"Analogical reasoning: king - man + woman = {analogy_result[0][0]}")
```

By performing these checks on your Word2Vec model, you can gain confidence in its success and usefulness for tasks involving word embeddings.

```python
# Assuming you have trained a Word2Vec model and stored it in the 'model' variable

word = 'anies'
similar_words = model.wv.most_similar(word)

print(f"Similar words to '{word}':")
for similar_word, similarity_score in similar_words:
    print(similar_word, similarity_score)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Assuming you have trained a Word2Vec model and stored it in the 'model' variable

# Get the word vectors and corresponding words
words = list(model.wv.key_to_index.keys())
word_vectors = model.wv[words]

# Reduce the dimensionality using t-SNE
tsne = TSNE(n_components=2, random_state=42)
word_vectors_tsne = tsne.fit_transform(word_vectors)

# Plot the word vectors in 2D space
plt.figure(figsize=(10, 10))
plt.scatter(word_vectors_tsne[:, 0], word_vectors_tsne[:, 1], marker='.')

# Add labels to the points
for i, word in enumerate(words):
    plt.annotate(word, (word_vectors_tsne[i, 0], word_vectors_tsne[i, 1]))

plt.show()
```

**is it a common thing if TSNE and UMP has different visual even they have a same dataset?**




Yes, it is common for t-SNE and UMAP to produce different visualizations even when applied to the same dataset. These algorithms use different mathematical approaches and optimization objectives, resulting in variations in how they capture and represent the underlying structure of the data.

Both t-SNE and UMAP are non-linear dimensionality reduction techniques that aim to project high-dimensional data into lower-dimensional spaces while preserving certain relationships between data points. However, they make different assumptions and employ different strategies to achieve this goal.

The differences in the visualizations can arise due to several factors, including:

1. Algorithmic Approach: t-SNE and UMAP use different algorithms and mathematical frameworks, which can lead to divergent results. UMAP focuses on preserving both local and global structure, while t-SNE primarily emphasizes local relationships.

2. Sensitivity to Parameters: Both algorithms have parameters that need to be tuned, such as the number of neighbors, perplexity, minimum distance, learning rate, and so on. The choice of parameter values can significantly impact the resulting visualizations.

3. Randomness: Both t-SNE and UMAP involve a degree of randomness during their initialization or optimization steps. Randomness can introduce variations in the computations and affect the final visualizations.

Given these factors, it is quite normal for t-SNE and UMAP to yield different visualizations even when applied to the same dataset. It is important to explore the parameter space, try different configurations, and interpret the visualizations with caution, considering the strengths and limitations of each algorithm.
